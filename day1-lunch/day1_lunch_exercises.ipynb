Plot multiple series

```python
x2 = [2, 4, 6]
y2 = [8, 64, 216]
```

Annotate plot

```python
fig, ax = plt.subplots()
ax.plot( x, y )
ax.plot( x2, y2, '+' )
plt.show()
```

Subplot 

```python
fig, ax = plt.subplots()
ax.plot( x, y, label="x^2" )
ax.plot( x2, y2, '+', label="x^3" )
ax.set_title( "Polynomials" )
ax.set_xlabel( "x" )
ax.set_ylabel( "f(x)" )
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots( nrows=2, sharex = True, sharey = True )
ax[0].plot( x, y, '.' )
ax[1].plot( x2, y2 )
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bed_columns = ["chr", "start", "end"]
bed_columns
```

```python
df_k4 = pd.read_table( "../../qbb2020/data/K4me3.bed",
                        names=bed_columns)
df_k4
```

```python
df_k4.shape
```

```python
df_k4.loc[:, "width"] = df_k4.loc[:,"end"] - df_k4.loc[:,"start"]
df_k4
```

```python
fig, ax = plt.subplots()
ax.hist( df_k4.loc[:,"width"])

plt.show()
```

```python
#extracting chr column 
chr_k4_regions = pd.Index(df_k4.loc[:, "chr"])

#counting frequencies in series
chr_k4 = chr_k4_regions.value_counts()
print(chr_k4)

sorted_chr_k4 = chr_k4.sort_index()
print(sorted_chr_k4)



fig, ax = plt.subplots()
ax.set_title("K4me3 Region Counts")
ax.set_ylabel("Counts")
ax.set_xlabel("Regions")
ax.bar(sorted_chr_k4.index[:-2], sorted_chr_k4[:-2] )
plt.show()
```

Plots for K9 and K27

Extracting bed files for K9/K27

```python
df_k9 = pd.read_table( "../../qbb2020/data/K9me3.bed",
                        names=bed_columns)
print(df_k9)

df_k27 = pd.read_table( "../../qbb2020/data/K27me3.bed",
                        names=bed_columns)
print(df_k27)
```

```python
Calculating width for chromosome regions
```

```python
df_k9.loc[:, "width"] = df_k9.loc[:,"end"] - df_k9.loc[:,"start"]
df_k9
```

```python
df_k27.loc[:, "width"] = df_k27.loc[:,"end"] - df_k27.loc[:,"start"]
df_k27
```

```python
#extracting chr_k9 column 
chr_k9_regions = pd.Index(df_k9.loc[:, "chr"])

#counting frequencies in series
chr_k9 = chr_k9_regions.value_counts()
print(chr_k9)

sorted_chr_k9 = chr_k9.sort_index()
print(sorted_chr_k9)



fig, ax = plt.subplots()
ax.set_title("K9me3 Region Counts")
ax.set_ylabel("Counts")
ax.set_xlabel("Regions")
ax.bar(sorted_chr_k9.index[:-2], sorted_chr_k9[:-2] )
plt.show()
```

```python
#extracting chr_k27 column 
chr_k27_regions = pd.Index(df_k27.loc[:, "chr"])

#counting frequencies in series
chr_k27 = chr_k27_regions.value_counts()
print(chr_k27)

sorted_chr_k27 = chr_k27.sort_index()
print(sorted_chr_k27)



```

Exercise 3: Creating chromosome distrubtion figure with 3 subplots

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize = (23, 6))
fig.suptitle("Chromosome Distributions")

#k4me3
ax1.bar(sorted_chr_k4.index[:-2], sorted_chr_k4[:-2] )
ax1.set_title("K4me3 Region Counts")
ax1.set_ylabel("Counts")
ax1.set_xlabel("Regions")

#k9me3
ax2.bar(sorted_chr_k9.index, sorted_chr_k9 )
ax2.set_title("K9me3 Region Counts")
ax2.set_ylabel("Counts")
ax2.set_xlabel("Regions")

#k27me3
ax3.bar(sorted_chr_k27.index, sorted_chr_k27 )
ax3.set_title("K27me3 Region Counts")
ax3.set_ylabel("Counts")
ax3.set_xlabel("Regions")
plt.show()
```
