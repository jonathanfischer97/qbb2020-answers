```python
import pandas as pd
```

```python
ref = open('mm10_refseq_genes_chr6_50M_60M.bed', 'r')
e4 = open('E4sorted.bedGraph')
e5 = open('E5sorted.bedGraph')

#extract genes and positions into dictionary
ref_dict = {}

for line in ref:
    gene=line.split('\t')
    start = gene[4]
    end = gene[5]
    name = gene[12]
    ref_dict[name] = (int(start),int(end))

#extract methylation sites into dictionary for E4 cells
e4_dict = {}

for line in e4:
    if 'chr' not in line:
        pass
    else:
        site = line.split('\t')
        start = site[1]
        meth = site[3]
        e4_dict[int(start)] = float(meth)

#extract methylation sites into dictionary for E5.5 cells 
e5_dict = {}

for line in e5:
    if 'chr' not in line:
        pass
    else:
        site = line.split('\t')
        start = site[1]
        meth = site[3]
        e5_dict[int(start)] = float(meth)


#loop through ref genes and check for methylation within 
enriched = {}

for gene in ref_dict:
    start=ref_dict[gene][0]
    end=ref_dict[gene][1]
    
    #initialize tallies for number of meth sites and total meth percentage for both E4 and E5.5
    sites4=0
    total4=0
    sites5=0
    total5=0
    
    for i in range(start,end):
        if i in e4_dict:
            sites4+=1
            total4+=e4_dict[i]
        if i in e5_dict:
            sites5+=1
            total5+=e5_dict[i]
    #calculate fold change from E4 to E5.5 using average methylation percentage per gene
    if total4!=0 and (total5/sites5)>(total4/sites4):
        enriched[gene] = (total5/sites5)/(total4/sites4)
```

```python
pd.DataFrame(list(enriched.items()),columns = ['Genes','Fold change'])
```
