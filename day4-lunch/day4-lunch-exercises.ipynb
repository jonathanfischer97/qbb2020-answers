```python
#preprocessed gtf file in UNIX to exclude header and non-3R 
fs = open('chr3R.gtf', 'r')
genome = fs.readlines()
fs.close()
```

```python
gene_list = []
list_size = 0 
for line in genome:
    line_split = line.split('\t')
    if "gene" in line_split[2] and "protein_coding" in line_split[8]:
        new_list = [line_split[3], line_split[4], line_split[-1]]
        gene_list.append(new_list)
        #list_size +=1
        #if list_size > 10:  #remove once tested
            #break 
print(gene_list)
```

```python
#initialize variables
search = 21378950
low = 0 
high = len(gene_list)-1
middle = int((high+low)/2)
previous_mid = None

tally = 0
BINGO = False 

while not BINGO:
    tally += 1
    if int(gene_list[middle][1]) < search:
        low = middle +1
    else: 
        high = middle 
    previous_mid = middle
    middle = int((high+low)/2)
    if previous_mid == middle:
        BINGO = True

        distance = abs(search - int(gene_list[middle][0]))
        last_column = gene_list[middle][2].split(';')

        print('gene: ' + str(last_column[0]) + ', ' + 'tally: ' + str(tally) + ', ' 'distance: '+ str(distance))




```
